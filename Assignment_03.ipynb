```python
operator = input("Enter Operator + or - or * or /")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter Second number: "))

if operator == '+':
    print("Result: ", num1 + num2)
elif operator == '-':
    print("Result: ", num1 - num2)
elif operator == '*':
    print("Result: ", num1 * num2)
elif operator == '/':
    print("Result: ", num1 / num2)
else:
    print("Error! ")
             
```
